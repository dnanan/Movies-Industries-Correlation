```python
# Importing libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
from matplotlib.pyplot import figure
import seaborn as sns

plt.style.use('ggplot')
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # adjust the configuration of the plots we will create
```

```python
# Reading & Looking the data
df = pd.read_csv('movies.csv')
df.head()
```

```python
# Looking for missing data
# Creating a for loop to loop through all the columns to find missing values
for col in df.columns:
  missing = np.mean(df[col].isnull())
  print('{} - {}%'.format(col, missing))
```

```python
# Dealing with missing values
df.fillna(value=0, inplace=True)
```

```python
# Looking for data types
#df.info()
df.dtypes
```

```python
# Changing the budget and gross column to int
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df = df.sort_values(by = ['gross'], inplace= False, ascending = False)
df.head()
```

```python
# Display all rows so i can have better understanding of the data
pd.set_option('display.max_rows',None)
```

```python
df['company'] = df['company'].astype('str')
```

```python
# droping any duplicates
df.drop_duplicates().head()
```

```python
df.shape
```

```python
# Budget high correlation
# company high correlation
```

```python
# Scatter plot with budget vs gross

plt.scatter(x = df['budget'], y = df['gross'])
plt.title('Budget vs Gross Earning')
plt.xlabel('Budget Earning')
plt.ylabel('Gross Earning')
plt.show()
```

```python
# Plot the budget vs Gross using seaborn
sns.regplot(x = 'budget', y = 'gross', data = df, scatter_kws= {"color": "red"}, line_kws= {"color":"blue"}) # We can see we have a positive correlation
```

```python
# Looking at correlation
df.corr(method ="pearson") # Pearson, Kendall, Spearman correlation
# From what I can see there is a high correlation between budget and gross
```

```python
corr_matrix =  df.corr(method ="pearson")
plt.title('Correlation Matrix')
sns.heatmap(corr_matrix, annot = True)
plt.show()
```

```python
# Numerizing the columns to find the correlation
df_num = df

for col_name in df_num.columns:
  if(df_num[col_name].dtypes == 'object'):
    df_num[col_name] = df_num[col_name].astype('category')
    df_num[col_name] = df_num[col_name].cat.codes
```

```python
df_num.head()
```

```python
corr_matrix =  df_num.corr(method ="pearson")
plt.title('Correlation Matrix')
sns.heatmap(corr_matrix, annot = True)
plt.show()
```

```python
correlation_matrix = df_num.corr()
corr_pairs = correlation_matrix.unstack()
corr_pairs
```
